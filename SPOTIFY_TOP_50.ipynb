```python

#import numpy for numerical computing
import numpy as np
#importing Pandas for data manipulation and analysis
import pandas as pd
#importing Plotly Express for interactive plotting
import plotly.express as px
#importing matplotlib for static plotting
import matplotlib.pyplot as plt
#importing seaborn for stastistical data visualization
import seaborn as sns
```

```python
#Read the csv files
df = pd.read_csv("/content/top_50_2023 (1).csv")
df.head()
```

```python
#Compute summary statistics for numerical columns in the DataFrame
description1 = df.describe()
# Print the summary statistics
print(description1)
```

```python
# Generate a histogram
fig = px.histogram(df, x="artist_name", title = 'Top 50 Singers').update_xaxes(categoryorder = 'total descending')
# Display the generated histogram
fig.show()
```

```python
# Define the list of audio features to include in the correlation analysis
track_attributes =['danceability','valence','energy','acousticness','instrumentalness','liveness']

# Compute the correlation matrix for the selected track attributes
correlation_matrix = df[track_attributes].corr()

# Set up the figure size for the heatmap plot
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix,annot =True, xticklabels = track_attributes,yticklabels= track_attributes)

# Customize the tick labels' rotation for better readability
plt.xticks(range(len(track_attributes)), track_attributes, rotation=45)
plt.yticks(range(len(track_attributes)), track_attributes, rotation=0)

#tiltle of heatmap
plt.title("Heatmap")

#display the heatmap
plt.show()
```

```python
#Plot valence against danceability from a DataFrame.

def plot_valence_danceability(df):


    # Extract relevant columns
    valence = df['valence']
    danceability = df['danceability']

    # Plot
    plt.figure(figsize=(8, 6))
    plt.scatter(danceability, valence, color='blue')

    # Add labels and title
    plt.xlabel('Danceability')
    plt.ylabel('Valence')
    plt.title('Valence vs Danceability')

    # Show plot
    plt.grid(True)
    plt.show()
plot_valence_danceability(df)

```
